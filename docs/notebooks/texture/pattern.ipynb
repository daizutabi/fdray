```python
from fdray import *

renderer = Renderer(width=800, height=600)


def render(texture1, texture2):
    sphere = Declare(Sphere(0, 1.2))
    cylinder = Declare(Cylinder("-z", "z", 1))
    box = Declare((Box(-1, 1) - Box(-1, 1).translate(1)).rotate("-15*z"))
    shapes = Union(
        Object(sphere).translate("-2.7*y"),
        Object(cylinder),
        Object(box).translate("2.7*y"),
    )
    camera = Camera(0, 20, view_scale=4)
    scene = Scene(
        Include("colors.inc"),
        camera,
        LightSource(camera.location, "white"),
        shapes.add(texture1).translate("1.5*z"),
        shapes.add(texture2).translate("-1.5*z"),
    )
    return renderer.render(scene, return_image=True)
```

```python
render(
    [Color("White*0.9"), Normal("agate", 1).scale(0.8)],
    [Color("White*0.9"), Normal("agate", 1).scale(0.8)],
)
```
