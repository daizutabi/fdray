```python
import numpy as np

from fdray import *
from fdray.data.texture import data_texture

# 温度データをシミュレート (2D配列)
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
temperature = np.sin(np.sqrt(X**2 + Y**2)) * np.exp(-0.1 * (X**2 + Y**2))

# 温度データからテクスチャを作成（赤-黄-白のカラーマップ）
temp_texture = data_texture(
    data=temperature,
    colormap="hot",  # "hot", "viridis", "plasma", "inferno"などから選択
    min_value=-0.5,  # データの最小値（省略可能）
    max_value=0.5,  # データの最大値（省略可能）
    scale=1.0,  # テクスチャのスケール
)

# テクスチャを球体に適用
sphere = Sphere((0, 0, 0), 1.0).add(temp_texture)
```

```python
Scene(
    Camera(0, 0, view_scale=2),
    LightSource(0, "white"),
    # Sphere((0, 0, 0), 1.0, Color("red")),
    sphere,
).render(400, 400)
```
