# Light source

## Point Lights

```python
from fdray import *

T = Declare(
    Texture(
        Pigment("brick", Color("white"), Color((0.8, 0.6, 0.4))).scale(0.1),
        Normal("brick", Normal("bumps", 0.2), Normal("granite", 0.3)).scale(0.1),
    ),
)
Scene(
    Camera(270, 10, look_at=(0, 0, 1), view_scale=2, distance=3),
    LightSource((0, 0, 5), "white", from_camera=False),
    Sphere("z", 0.4, Color((0.5, 0.7, 0.7))),
    Box(-1, 1).scale(3, 5, 1).translate(0, 0, -1).add(Texture(T)),
    Box(-1, 1).scale(1, 5, 5).translate(-3, 0, 4).add(Texture(T).rotate("x*90")),
    Box(-1, 1).scale(1, 5, 5).translate(3, 0, 4).add(Texture(T).rotate("x*90")),
    Box(-1, 1).scale(3, 1, 5).translate(0, 5, 4).add(Texture(T).rotate("x*90")),
).render(300, 300)
```
