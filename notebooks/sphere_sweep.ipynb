```python
import numpy as np
from PIL import Image

from fdray import (
    Background,
    Camera,
    Color,
    Curve,
    LightSource,
    Polyline,
    Renderer,
    Scene,
)
```

```python
def scene(*attrs) -> Scene:
    camera = Camera(0, 20, view_scale=1.4, look_at=(0, 0.5, 0.5))
    return Scene(
        camera,
        LightSource(camera.location, "white"),
        Background("white", alpha=0.1),
        *attrs,
    )


def render(*attrs) -> Image.Image:
    renderer = Renderer(width=300, height=300)
    array = renderer.render(scene(*attrs))
    return Image.fromarray(array)
```

```python
a = np.array([[0, 0, 0], [0, 0, 1], [0, 1, 1], [0, 1, 0]])
x = Polyline(a, 0.1, Color("red"))
render(x)
```

```python
x = Curve(a, 0.1, Color("red"))
render(x)
```
